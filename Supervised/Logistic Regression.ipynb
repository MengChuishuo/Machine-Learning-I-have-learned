```python
class LogisticRegressionClassifier:
    def __init__(self, max_iter=100, learning_rate=0.001, reg_coef=0.01, initial_weights=None):
        self.max_iter = max_iter
        self.learning_rate = learning_rate
        self.reg_coef = reg_coef
        self.weights = initial_weights.copy() if initial_weights is not None else None

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def data_matrix(self, X):
        X = np.array(X)
        return np.hstack([np.ones((X.shape[0], 1)), X])

    def fit(self, X, y):
        data_mat = self.data_matrix(X)
        n_samples, n_features = data_mat.shape
        
        if self.weights is None:
            self.weights = np.zeros((n_features, 1), dtype=np.float32)
        
        for iter_ in range(self.max_iter):
            for i in range(n_samples):
                x_i = data_mat[i].reshape(-1, 1)  
                y_i = y[i]
                
                pred = self.sigmoid(np.dot(x_i.T, self.weights))
                error = y_i - pred
                reg_term = self.reg_coef * self.weights
                reg_term[0] = 0  
                
                self.weights += self.learning_rate * (error * x_i - reg_term)
                
    def predict(self, X):
        data_mat = self.data_matrix(X)
        scores = np.dot(data_mat, self.weights)
        return (scores > 0).astype(int).flatten()
```
